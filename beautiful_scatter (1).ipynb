```python
import matplotlib
import matplotlib.pyplot as plt #import matplotlib and pyplot as plt

x_values = range(1,5000) # define x values for plot
y_values = [x**2 for x in x_values] #list comprehension to make squared values

fig, ax = plt.subplots() #make plot

ax.set_xlabel("Values", fontsize=14)
ax.set_ylabel("Cubes", fontsize=14)
ax.set_title("First 5000 cubes", fontsize=20)  #add labels and title, can adjust size!

ax.scatter(x_values, y_values, c = y_values, cmap = plt.cm.Blues, linewidth=3) #make scatter graph with coluring gradient

ax.tick_params(labelsize=14)
ax.ticklabel_format(style = 'plain') # last few adjustments to labels

ax.axis([0, 5500, 0, 27500000]) #axis values


plt.show() #plot scatter graph
```
